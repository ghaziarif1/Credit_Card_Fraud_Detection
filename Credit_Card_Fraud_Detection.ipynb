Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
predict_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
predict_card_data['V26'].fillna(predict_card_data['V26'].median(), inplace=True)
predict_card_data['V27'].fillna(predict_card_data['V27'].median(), inplace=True)
predict_card_data['V28'].fillna(predict_card_data['V28'].median(), inplace=True)
predict_card_data['Amount'].fillna(predict_card_data['Amount'].median(), inplace=True)
predict_card_data['Class'].fillna(predict_card_data['Class'].median(), inplace=True)
```

```python
predict_card_data.isnull().sum()
```

```python
#distribution of legit transcations & fraudulent transcations
predict_card_data['Class'].value_counts()
```

This  dataset is highly unbalanced

0 -> Normal Transcation

1 -> Fraudulent Transcation

```python
#separating the data for analysis
legit =  predict_card_data[predict_card_data.Class == 0]
fraud = predict_card_data[predict_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transcations
predict_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal transcations and fraudulent transcations

```python
legit_sample = legit.sample(n=3)
```

Concatenationg 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
